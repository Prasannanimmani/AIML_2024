```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

df['target'] = cancer.target

print(df.info())

print(df.isnull().sum())

print("\nTarget class distribution:")
print(df['target'].value_counts())

sns.boxplot(
    data=df.iloc[:, :10],
    orient="v",
    showmeans=True
)
plt.title("Boxplot of First 10 Features in Breast Cancer Dataset")
sns.plt.show()

corr = df.corr()
sns.heatmap(corr, annot=True)
sns.plt.title("Correlation Matrix")
sns.plt.show()

sns.pairplot(df.iloc[:, :-1], hue="target")
sns.plt.suptitle("Pairplots of Features with Target Class")
sns.plt.show()
```
